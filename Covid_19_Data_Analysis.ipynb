```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('country_wise_latest.csv')
```

```python
df.info()
print(df.isnull().sum())
df=df.dropna()
```

```python
df=df[df['Country/Region']=='Bangladesh']
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index,df['Confirmed'],label='Confirmed Cases',color='blue',marker='o')
plt.xlabel("Index")
plt.ylabel("Confirmed Cases")
plt.title("Covid 19 Confirmed Cases over Index")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index,df['Deaths'],label='Deaths',color='red',marker='o')
plt.plot(df.index,df['Recovered'],label='Recovered',color='green',marker='o')
plt.xlabel("Index")
plt.ylabel("Count")
plt.title("Covid 19 Deaths and Recoveries over Index")
plt.legend()
plt.grid()
plt.show()
```

```python
df['Death Rate (%)']=(df['Deaths']/df['Confirmed'])*100
plt.figure(figsize=(10,5))
plt.plot(df.index,df['Death Rate (%)'],label='Death Rate (%)',color='purple',marker='o')
plt.xlabel("Index")
plt.ylabel("Death Rate (%)")
plt.title("Covid 19 Deaths Rate over Index")
plt.legend()
plt.grid()
plt.show()
```

```python
df.to_csv('cleaned_covid_data.csv')
```
